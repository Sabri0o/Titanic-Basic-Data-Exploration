```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
titanic = pd.read_csv(open('titanic_data.csv'))
print (titanic.shape)
print (titanic.head())
print (titanic.dtypes)

print ('*******************************')

titanic = titanic.drop(['PassengerId','Ticket','Cabin','Embarked','Name'], axis=1)
print (titanic['Age'].shape)
#Return number of non-NA/null observations in the Series
print (titanic['Age'].count())
# filter out rows based on missing values in a the column 'Age'
titanic = titanic.loc[titanic['Age'].notnull()]

print (titanic.shape)
```

```python

def age_cat(x):
    
    if x >=18:
        return 'Adult'
    if x >= 11 and x < 18:
        return 'Teenager'
    if x >= 2 and x < 11:
        return 'Child'
    if x > 0 and x < 2:
        return 'Baby'

def Yes_or_No(x):
    if x ==0:
        return 'No'
    if x ==1:
        return 'Yes'
    

titanic['Age_cat']=titanic['Age'].copy().apply(age_cat)
titanic['Survived']=titanic['Survived'].apply(Yes_or_No)

#print (titanic['Age_cat'].value_counts())
#print (titanic['Age_cat'].head())
fig =plt.figure(figsize=(8,8))
sns.countplot(x="Pclass",hue="Age_cat", data=titanic)
sns.despine()    
    
```

```python
def survival_rate(data,sex,socialclass,age_cat):
    if age_cat in data.loc[(data['Sex']==sex) & (data['Pclass']==socialclass),'Age_cat'].values:
        return(data.loc[(data['Survived']=='Yes') & (data['Sex']==sex) & (data['Pclass']==socialclass) & (data['Age_cat']==age_cat),'Survived'].count()/
           data.loc[(data['Sex']==sex) & (data['Pclass']==socialclass) & (data['Age_cat']==age_cat),'Survived'].count()*100).round(2)

for sex in ['female','male']:
    for socialclass in [1,2,3]:
        for age_cat in ['Adult','Teenager','Child','Baby']:
            print(sex,age_cat,'class',socialclass,'survival rate:',survival_rate(titanic,sex,socialclass,age_cat))
```
